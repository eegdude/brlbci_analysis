```python
%load_ext autoreload
%autoreload 2
import folders
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [6,6]
sns.set(font_scale=3, palette='deep') 
sns.set_style("white")
```

```python
markup=pd.read_csv(folders.markup_path)
subset = markup.loc[markup['reg']=='brl_static6_all8']
blind_subset = subset.loc[subset['blind']==1]
```

```python
# demographics - age and sex
with sns.color_palette("colorblind"):
    g = sns.FacetGrid(subset, col="blind", hue="sex", col_wrap=2, height=12, legend_out=True)    
    g.map(sns.distplot, 'age', kde=False, bins=[18, 25, 35, 45, 60])
    g.add_legend()

```

```python
# demographics - years of vision loss
with sns.color_palette("colorblind"):
    g = sns.FacetGrid(blind_subset,  col="blind", hue="congenitaly_blind", col_wrap=1, height=12, legend_out=True)    
    g.map(sns.distplot, 'blind_years', kde=False, bins=[5,10,15,20,25,30,35,40,45,50])
    g.add_legend()
    plt.yticks([0,1,2])
    plt.xlabel('Years after vision loss')
    new_labels = ['Congenital blindness', 'Acquired blindness']
    for t, l in zip(g._legend.texts, new_labels): t.set_text(l)
```
